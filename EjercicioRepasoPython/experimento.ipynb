```python
epsilon = 0.07
min_pts = 15
nuclei,frontier=[],[]
def DBScan(data):
    num1,num2,num3=[],[],[]
    for i in range(0,len(data)):
        z=np.sqrt((data['Latitud'][i]-data['Latitud'])**2+(data['Longitud'][i]-data['Longitud'])**2)
        z=z<=epsilon
        if z.sum()-1<min_pts and z.sum()-1>=0:
            num1.append(i)
        elif (z.sum()-1==0 or z.sum()-1>=min_pts):
            num2.append(i)
    frontier,nucleir=data.drop(num2,axis=0),data.drop(num1,axis=0)
    frontier,nucleir=frontier.reset_index(drop=True),nucleir.reset_index(drop=True)
    for i in range(0,len(frontier)):#candidatos en x a ser frontera
        b=np.sqrt((frontier['Latitud'][i]-nucleir['Latitud'])**2+(frontier['Longitud'][i]-nucleir['Longitud'])**2)
        b=b<=epsilon
        if b.sum()==0:
            num3.append(i)
    frontier=frontier.drop(num3,axis=0)
    return nucleir,frontier
nucleo,frontera=DBScan(data)
num_outliers = len(nucleo)+len(frontera)-len(data)
```

```python
def k_vecinos(data,clases,k=4):
    ind_or=list(data.index)
    grupo = []
    
    for i in ind_or:
        aux = []
        d = np.sqrt( (data['R'][i] - clases['R'])**2 + (data['G'][i] - clases['G'])**2 + (data['B'][i] - clases['B'])**2 )
        for j in range(0,k):
            m, o, az = min(d), d.index, list(d)
            ind_min = az.index(m)
            ind_min_d = o[ind_min]
            d=d.drop(ind_min_d)
            aux.append(x['Tribu'][ind_min_d])
        d = [aux.count('A'),aux.count('B'),aux.count('C')]
        fin=d.index(max(d))
        if fin ==0:
            grupo.append('A')
        elif fin ==1:
            grupo.append('B')
        else:
            grupo.append('C')
            
    return grupo
```

```python
#clase de la capa de la red

class neural_layer():
    
    def __init__(self,n_conn,n_neur,act_f):
        self.act_f = act_f
        
        self.b = np.random.rand(1, n_neur) * 2 -1
        
        self.w = np.random.rand(n_conn, n_neur) * 2 -1
```

```python
# funciones de activacion

sigm = (lambda x: 1/(1 + np.e ** (-x)),
       lambda x: x * (1-x))

_x=np.linspace(-5,5,100)
plt.plot(_x, sigm[0](_x))
```

```python

#esta funcion crea de forma automatica la red
#Solo necesita un vector con la estructura de la red y la funcion de activacion
def create_nn(topology, act_f):
    
    #Aqui va la red neuoronal:
    nn = []
    
    #Este for crea una capa por cada iteracion
    for l, layer in enumerate(topology[:-1]):
        
        #agregamos la capa a la red,
        nn.append(neural_layer(topology[l],topology[l+1], act_f))
        
    return nn

```
